### Desafio: Valores da média, mediana, moda e desvio padrão da pontuação de crédito para cada estado do dataset

```python
cliente_estado = clientes.groupby(['estado_residencia'])
cliente_estado.head(5)
```

```python
df_estado = clientes[['pontuacao_credito', 'estado_residencia']]
cliente_estado['media'] = df_estado.groupby(['estado_residencia']).mean()
cliente_estado['mediana'] = df_estado.groupby(['estado_residencia']).median()
cliente_estado['desvio_padrao'] = df_estado.groupby(['estado_residencia']).std()
cliente_estado.head(5)
```

#### Como o .mode() retorna um dataframe, devemos tratá-lo de forma separada

```python
cliente_estado['moda'] = df_estado.groupby(['estado_residencia']).agg(lambda x:x.value_counts().index[0])
cliente_statistics = cliente_estado.drop(['pontuacao_credito', 'mean'], axis=1)
cliente_statistics = cliente_statistics[['moda', 'mediana', 'media', 'desvio_padrao']]
cliente_statistics
```

#### Conversão dos tipos nas colunas moda e mediana para float

```python
cliente_statistics.moda = cliente_statistics.moda.astype(float)
cliente_statistics.mediana = cliente_statistics.mediana.astype(float)
cliente_statistics.dtypes()
```

### Convertendo para JSON

```python
cliente_statistics.to_json(r'submission2.json', orient='index')
```
